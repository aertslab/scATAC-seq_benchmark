# do for all dars

```bash
LC_ALL=C 

for bed in downstream_analysis/DARs/*/*__DARs__unique.bed
do
    echo $bed
    newbed=${bed%unique.bed}unique.sorted.bed
    echo $newbed
    cat $bed | sort --parallel=8 -S 8G -k 1,1 -k 2,2n -k 3,3n > $newbed
done
# LC_ALL=C sort --parallel=8 -S 8G -k 1,1 -k 2,2n -k 3,3n
```

```bash
for bed in downstream_analysis/DARs/*/*__DARs__unique.sorted.summits.bed
do
    echo $bed
    newbed=${bed%.sorted.summits.bed}.sorted.summits.nearesttssdist.bed
    bedtools closest -nonamecheck -d -t first -a $bed -b tss.sorted.bed > $newbed &
done
# LC_ALL=C sort --parallel=8 -S 8G -k 1,1 -k 2,2n -k 3,3n
```

# do for all fragments files

```bash
module load mawk
```

```bash
for FILEPATH in ../1_data_repository/fixedcells_fragments/*.FIXEDCELLS.fragments.tsv.gz
do
    echo $FILEPATH
    OUTPATH=${FILEPATH%.fragments.tsv.gz}.fragments.bed
    echo $OUTPATH
    zcat $FILEPATH | mawk -F "\t" '{print $1 "\t" $2 "\t" $3 "\t" NR}' > $OUTPATH &
done
```

now calculate nearest TSS

```bash
LC_ALL=C 

for bed in ../1_data_repository/fixedcells_fragments/*.FIXEDCELLS.fragments.bed
do
    echo $bed
    newbed=${bed%.bed}.sorted.nearesttssdist.bed
    sort -k1,1 -k2,2n $bed > ${bed%.bed}.sorted.bed & 
done
# LC_ALL=C sort --parallel=8 -S 8G -k 1,1 -k 2,2n -k 3,3n
```
