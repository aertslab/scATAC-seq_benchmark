```python
import seaborn as sns

sns.set_context("notebook")
sns.set_style("white")
```

```python
fig, axes = plt.subplots(8, 6, figsize=(12, 16))
axes = axes.flatten()
for sample, path in cell_data_path_dict.items():
    print(sample, path)
    print(all_order.index(sample))
    df = pd.read_csv(path, index_col=0, sep="\t")
    ax = axes[all_order.index(sample)]
    sns.scatterplot(
        data=df,
        x="tSNE_1",
        y="tSNE_2",
        hue="consensus_cell_type",
        palette=cell_type_color_dict,
        s=3,
        linewidth=0,
        ax=ax,
    )
    sns.despine(
        left=True,
        bottom=True,
        right=True,
        top=True,
    )
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.set_xlabel(None)
    ax.set_ylabel(None)
    ax.set_title(sample_id_ultrashort_alias_dict[sample])

    ax.legend().remove()

axes[-1].set_xticklabels([])
axes[-1].set_yticklabels([])
axes[-1].set_xlabel(None)
axes[-1].set_ylabel(None)

plt.savefig("plts_final/all_tsne.png", dpi=600, transparent=True)
plt.show()
plt.close()
```
