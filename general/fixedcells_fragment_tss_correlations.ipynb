```python
profile_data_dict = {
    x.split("/")[-1].split("__")[0]: x
    for x in sorted(
        glob.glob(
            "../fixedcells_3_cistopic_consensus/cistopic_qc_out_CONSENSUS/*profile_data.pkl"
        )
    )
}
```

```python
df_merged = pd.DataFrame()

for sample, path in profile_data_dict.items():
    print(sample)
    with open(path, "rb") as f:
        profile_data = pickle.load(f)
        tech = sample.split("_")[1]
        df = profile_data["profile_tss"]
        df["sample"] = sample
        df["tech"] = tech
        df_merged = pd.concat([df_merged, df], axis=0, ignore_index=True)

plt.show(plt.close())
```

```python
sns.lineplot(df_merged, x="Position", y="TSS_enrichment", hue="tech")
```

```python
df_merged_len = pd.DataFrame()

for sample, path in profile_data_dict.items():
    print(sample)
    with open(path, "rb") as f:
        profile_data = pickle.load(f)
        tech = sample.split("_")[1]
        df = profile_data["insert_size_distribution"]
        df["sample"] = sample
        df["tech"] = tech
        df_merged_len = pd.concat([df_merged_len, df], axis=0, ignore_index=True)
```
