```python
n_cells_dict = {
    "B cell": 555,
    "CD14+ monocyte": 747,
    "CD16+ monocyte": 144,
    "CD4+ T cell": 1008,
    "Cytotoxic T cell": 1769,
    "Dendritic cell": 83,
    "Natural killer cell": 126,
}
```

```python
random.seed(10)
```

```python
bcs_merged = []
for cell_type in cto_all.cell_data["harmony_consensus_cell_type__mega"].unique():
    for tech in cto_all.cell_data["tech"].unique():
        df_tmp = cto_all.cell_data[
            (cto_all.cell_data["tech"] == tech)
            & (cto_all.cell_data["harmony_consensus_cell_type__mega"] == cell_type)
        ]
        bcs = random.sample(list(df_tmp.index), n_cells_dict[cell_type])
        print(len(bcs))
        bcs_merged = bcs_merged + bcs
```

```python
len(bcs_merged)
```

```python
cto_subset = cto_all.subset(cells=bcs_merged, copy=True)
```

```python
cto_subset.cell_data.groupby(["harmony_consensus_cell_type__mega"])[
    "tech"
].value_counts()
```
