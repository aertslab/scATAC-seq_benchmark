```python
import glob
import pandas as pd
import pickle
import os
```

```python
metadata_path_dict = {x.split('/')[-1].split('__')[0]: x for x in sorted(glob.glob('cistopic_qc_out/*metadata_bc.pkl'))}
metadata_path_dict
```

```python
cto_path_dict = {x.split('/')[-1]: x for x in sorted(glob.glob('cistopic_objects/*cto*.pkl'))}
cto_path_dict
```

```python
metadata_path_dict
```

```python
write = True
aberrant_samples = []
aberrant_ctos = []
for cto_name in cto_path_dict.keys():
    cto_path = cto_path_dict[cto_name]
    print(cto_path)
    with open(cto_path, 'rb') as f:
        cto = pickle.load(f)
        
    sample = cto_name.split('__')[0]
    with open(metadata_path_dict[sample], 'rb') as f:
        metadata_bc_df = pickle.load(f)
    
    if sample in ['VIB_hydrop_1.FIXEDCELLS', 'VIB_hydrop_2.FIXEDCELLS']:
        continue
        # metadata_bc_df.index = [x.split('___')[0] + '-' + x.split('___')[1].split('.')[-2][-1] + '___' + x.split('___')[1].replace('11','1').replace('12','2') for x in cto.cell_data.index]
        # metadata_bc_df.index = [x.replace('-2','').replace('-1','').replace('_1','_11').replace('_2','_12') for x in metadata_bc_df.index]
    # elif sample == 'VIB_hydrop_'
        
    df_sub = metadata_bc_df.loc[list(cto.cell_data.index)]
    for col in df_sub.columns:
        diff = cto.cell_data[col].compare(df_sub[col])
        if len(diff) == 0:
            print(f"{col} equal")
        else:
            print(f"{col} not equal")
            print(diff)
            if col in ["TSS_enrichment", "Log_total_nr_frag equal", "Log_unique_nr_frag", "sample_id"]:
                cto.cell_data[col] = df_sub[col]
                    
            else:
                print(f"{col} aberrant!!!")
                aberrant_samples.append(sample)
                aberrant_ctos.append(cto_path)
                
            if write == True:
                with open(
                    cto_path, "wb"
                ) as f:
                    pickle.dump(cto, f, protocol=4)                    
    print("\n")
```
