```python
import pandas as pd

%load_ext lab_black
```

```python
df_cell_counts = pd.read_csv("cell_counts.csv", index_col=0)
df_cell_counts
```

```python
df_cell_counts["rna_exclusive"] = (
    df_cell_counts["rna_cells"] - df_cell_counts["common_cells"]
)
df_cell_counts["atac_exclusive"] = (
    df_cell_counts["atac_cells"] - df_cell_counts["common_cells"]
)
```

```python
# library
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

for sample in df_cell_counts.index:
    venn2(
        subsets=(
            df_cell_counts.at[sample, "rna_exclusive"],
            df_cell_counts.at[sample, "atac_exclusive"],
            df_cell_counts.at[sample, "common_cells"],
        ),
        set_labels=("RNA Cells", "ATAC Cells"),
    )
    plt.savefig(f"plots/{sample}_venn.svg", dpi=300, facecolor="white")
    plt.show()
```
