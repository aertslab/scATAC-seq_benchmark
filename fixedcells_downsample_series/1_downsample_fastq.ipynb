We need to downsample based on cell counts and a given read depth.

````python
import pandas as pd
import glob
import os
````

````python
df_stats_full = pd.read_csv('../general/full.general_statistics.tsv', sep='\t', index_col=0)
````

````python
df_stats_full
````

````python
df_stats_fixedcells = pd.read_csv('../general/libds.general_statistics.tsv', sep='\t', index_col=0)
````

````python
df_stats_fixedcells
````

````python
min_value = 35000
````

````python
downsampling_depth = [min_value] * len(df_stats_fixedcells["RPC"]) / df_stats_fixedcells["RPC"]
````

````python
downsampling_depth = downsampling_depth.dropna()
````

````python
for item in downsampling_depth.index:
    if downsampling_depth[item] >= 1:
        downsampling_depth = downsampling_depth.drop(item)
````

````python
downsampling_depth
````

````python
import os
````

````python
min_value_str = str(min_value).replace('000','k')

parallel_filename = f"downsample.{min_value_str}.parallel"
with open(parallel_filename, 'w') as f:
    for sample in downsampling_depth.index:
        rate = downsampling_depth[sample]
        if rate <= 1.0:
            if sample.split('_')[1] == "ddseq":
                reads = [1, 2]
            else:
                reads = [1, 2, 3]

            for read in reads:
                fastq = f"../../1_data_repository/libds_fastq/{sample}__R{str(read)}.LIBDS.fastq.gz"
                newfastq = f"{min_value_str}_fastq/{sample}__R{str(read)}.{min_value_str}.fastq.gz"
                if not os.path.exists(newfastq):
                    f.write(f"seqtk sample -s 111 {fastq} {rate} | bgzip -@ 64 > {newfastq}\n")
                    print(f"seqtk sample -s 111 {fastq} {rate} | bgzip -@ 64 > {newfastq}")

                else:
                    f.write(f"#{newfastq} already exists!\n")
                    print(f"#{newfastq} already exists!")
````

# check depths

````python
cell_counts = {}
import pickle
````

````python
for file in sorted(glob.glob('selected_barcodes/*.pkl')):
    print(file)
    sample = file.split('/')[-1].split('.')[0]
    with open(file, "rb") as f:
        bcs = pickle.load(f)
    
    cell_counts[sample] = len(bcs)
````

````python
df = pd.DataFrame()
for file in glob.glob('*k/R1_lengths.txt'):
    print(file)
    depth = file.split('/')[0]
    print(depth)
    file2 = file.replace('R1', 'R2')
    file3 = file.replace('R1', 'R3')

    df = pd.read_csv(file, header=None, sep='\t', index_col=0)
    df.index = [x.split('__')[0] for x in  df.index]
    df2 = pd.read_csv(file2, header=None, sep='\t', index_col=0)
    df2.index = [x.split('__')[0] for x in  df2.index]
    df3 = pd.read_csv(file3, header=None, sep='\t', index_col=0)
    df3.index = [x.split('__')[0] for x in  df3.index]
    
    df[2] = df2
    df[3] = df3
    counts_list = []
    for subset in df.index:
        sample = subset.split('/')[-1]
        cell_count = cell_counts[sample]
        counts_list.append(cell_count)
    df['count'] = counts_list
    df['RPC'] = df[1]/df['count']
    print(df)
````
