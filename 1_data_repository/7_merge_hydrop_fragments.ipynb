# Merge HyDrop fragments files
They are from the same experiment, but the emulsion was split in half and separately indexed. At the time we thought this could reduce barcode collision rate with the new design (384 x 384), but in hindsight this was not necessary, and all other HyDrop-ATAC experiments have been done without this splitting.

```bash
module load HTSlib
```

```bash
zcat libds_fragments/VIB_hydrop_11.LIBDS.fragments.tsv.gz | mawk -F '\t' '{print $1 "\t" $2 "\t" $3 "\t" $4 "-1\t" $5}' | bgzip >  libds_fragments/VIB_hydrop_11.LIBDS.fragments.sampleid.tsv.gz &
zcat libds_fragments/VIB_hydrop_12.LIBDS.fragments.tsv.gz | mawk -F '\t' '{print $1 "\t" $2 "\t" $3 "\t" $4 "-2\t" $5}' | bgzip >  libds_fragments/VIB_hydrop_12.LIBDS.fragments.sampleid.tsv.gz &
zcat libds_fragments/VIB_hydrop_21.LIBDS.fragments.tsv.gz | mawk -F '\t' '{print $1 "\t" $2 "\t" $3 "\t" $4 "-1\t" $5}' | bgzip >  libds_fragments/VIB_hydrop_21.LIBDS.fragments.sampleid.tsv.gz &
zcat libds_fragments/VIB_hydrop_22.LIBDS.fragments.tsv.gz | mawk -F '\t' '{print $1 "\t" $2 "\t" $3 "\t" $4 "-2\t" $5}' | bgzip >  libds_fragments/VIB_hydrop_22.LIBDS.fragments.sampleid.tsv.gz &
```

```bash
zcat fixedcells_fragments/VIB_hydrop_11.FIXEDCELLS.fragments.tsv.gz | mawk -F '\t' '{print $1 "\t" $2 "\t" $3 "\t" $4 "-1\t" $5}' | bgzip >  fixedcells_fragments/VIB_hydrop_11.FIXEDCELLS.fragments.sampleid.tsv.gz &
zcat fixedcells_fragments/VIB_hydrop_12.FIXEDCELLS.fragments.tsv.gz | mawk -F '\t' '{print $1 "\t" $2 "\t" $3 "\t" $4 "-2\t" $5}' | bgzip >  fixedcells_fragments/VIB_hydrop_12.FIXEDCELLS.fragments.sampleid.tsv.gz &
zcat fixedcells_fragments/VIB_hydrop_21.FIXEDCELLS.fragments.tsv.gz | mawk -F '\t' '{print $1 "\t" $2 "\t" $3 "\t" $4 "-1\t" $5}' | bgzip >  fixedcells_fragments/VIB_hydrop_21.FIXEDCELLS.fragments.sampleid.tsv.gz &
zcat fixedcells_fragments/VIB_hydrop_22.FIXEDCELLS.fragments.tsv.gz | mawk -F '\t' '{print $1 "\t" $2 "\t" $3 "\t" $4 "-2\t" $5}' | bgzip >  fixedcells_fragments/VIB_hydrop_22.FIXEDCELLS.fragments.sampleid.tsv.gz &
```

```bash
zcat full_fragments/VIB_hydrop_11.FULL.fragments.tsv.gz | mawk -F '\t' '{print $1 "\t" $2 "\t" $3 "\t" $4 "-1\t" $5}' | bgzip >  full_fragments/VIB_hydrop_11.FULL.fragments.sampleid.tsv.gz &
zcat full_fragments/VIB_hydrop_12.FULL.fragments.tsv.gz | mawk -F '\t' '{print $1 "\t" $2 "\t" $3 "\t" $4 "-2\t" $5}' | bgzip >  full_fragments/VIB_hydrop_12.FULL.fragments.sampleid.tsv.gz &
zcat full_fragments/VIB_hydrop_21.FULL.fragments.tsv.gz | mawk -F '\t' '{print $1 "\t" $2 "\t" $3 "\t" $4 "-1\t" $5}' | bgzip >  full_fragments/VIB_hydrop_21.FULL.fragments.sampleid.tsv.gz &
zcat full_fragments/VIB_hydrop_22.FULL.fragments.tsv.gz | mawk -F '\t' '{print $1 "\t" $2 "\t" $3 "\t" $4 "-2\t" $5}' | bgzip >  full_fragments/VIB_hydrop_22.FULL.fragments.sampleid.tsv.gz &
```

```bash
cat libds_fragments/VIB_hydrop_11.LIBDS.fragments.sampleid.tsv.gz libds_fragments/VIB_hydrop_12.LIBDS.fragments.sampleid.tsv.gz > libds_fragments/VIB_hydrop_1.LIBDS.unsorted.fragments.tsv.gz &
cat libds_fragments/VIB_hydrop_21.LIBDS.fragments.sampleid.tsv.gz libds_fragments/VIB_hydrop_22.LIBDS.fragments.sampleid.tsv.gz > libds_fragments/VIB_hydrop_2.LIBDS.unsorted.fragments.tsv.gz &
```

```bash
cat fixedcells_fragments/VIB_hydrop_11.FIXEDCELLS.fragments.sampleid.tsv.gz fixedcells_fragments/VIB_hydrop_12.FIXEDCELLS.fragments.sampleid.tsv.gz > fixedcells_fragments/VIB_hydrop_1.FIXEDCELLS.unsorted.fragments.tsv.gz &
cat fixedcells_fragments/VIB_hydrop_21.FIXEDCELLS.fragments.sampleid.tsv.gz fixedcells_fragments/VIB_hydrop_22.FIXEDCELLS.fragments.sampleid.tsv.gz > fixedcells_fragments/VIB_hydrop_2.FIXEDCELLS.unsorted.fragments.tsv.gz &
```

```bash
cat full_fragments/VIB_hydrop_11.FULL.fragments.sampleid.tsv.gz full_fragments/VIB_hydrop_12.FULL.fragments.sampleid.tsv.gz > full_fragments/VIB_hydrop_1.FULL.unsorted.fragments.tsv.gz &
cat full_fragments/VIB_hydrop_21.FULL.fragments.sampleid.tsv.gz full_fragments/VIB_hydrop_22.FULL.fragments.sampleid.tsv.gz > full_fragments/VIB_hydrop_2.FULL.unsorted.fragments.tsv.gz &
```

```bash
zcat libds_fragments/VIB_hydrop_1.LIBDS.unsorted.fragments.tsv.gz | LC_ALL=C sort --parallel=8 -S 8G -k 1,1 -k 2,2n -k 3,3n | bgzip > libds_fragments/VIB_hydrop_1.LIBDS.fragments.tsv.gz &
zcat libds_fragments/VIB_hydrop_2.LIBDS.unsorted.fragments.tsv.gz | LC_ALL=C sort --parallel=8 -S 8G -k 1,1 -k 2,2n -k 3,3n | bgzip > libds_fragments/VIB_hydrop_2.LIBDS.fragments.tsv.gz &
zcat full_fragments/VIB_hydrop_1.FULL.unsorted.fragments.tsv.gz | LC_ALL=C sort --parallel=8 -S 8G -k 1,1 -k 2,2n -k 3,3n | bgzip > full_fragments/VIB_hydrop_1.FULL.fragments.tsv.gz &
zcat full_fragments/VIB_hydrop_2.FULL.unsorted.fragments.tsv.gz | LC_ALL=C sort --parallel=8 -S 8G -k 1,1 -k 2,2n -k 3,3n | bgzip > full_fragments/VIB_hydrop_2.FULL.fragments.tsv.gz &
```

```bash
zcat fixedcells_fragments/VIB_hydrop_1.FIXEDCELLS.unsorted.fragments.tsv.gz | LC_ALL=C sort --parallel=8 -S 8G -k 1,1 -k 2,2n -k 3,3n | bgzip > fixedcells_fragments/VIB_hydrop_1.FIXEDCELLS.fragments.tsv.gz &
zcat fixedcells_fragments/VIB_hydrop_2.FIXEDCELLS.unsorted.fragments.tsv.gz | LC_ALL=C sort --parallel=8 -S 8G -k 1,1 -k 2,2n -k 3,3n | bgzip > fixedcells_fragments/VIB_hydrop_2.FIXEDCELLS.fragments.tsv.gz &
```
