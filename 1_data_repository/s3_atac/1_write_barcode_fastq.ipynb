```bash
sci_fastq_barcode_to_new_fastq() {
    mawk '{
        if( NR % 4 == 1) {
            delim_pos = index($1,":")
            print $1
            print substr($1,2,delim_pos-2)
            print "+"
            print "AAAAAAAAAAAAAAAAAAAAAAAAAA"
        }
    }'
}
```

```bash
fix_header() {
    mawk '{
        if( NR % 4 == 1) {
            delim_pos = index($1,":")
            print substr($1,0,delim_pos-1)
            next
        } else {
            print $1
            next
        }
    }'
}
```
