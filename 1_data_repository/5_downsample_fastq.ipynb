# Downsample all the FASTQ files to same depth per cell
We need to downsample based on cell counts and a given read depth.

````python
import pandas as pd
import glob
````

````python
df_stats = pd.read_csv('../general/full.general_statistics.tsv', sep='\t', index_col=0)
````

````python
df_stats
````

````python
df_stats
````

````python
df_stats['RPC'].sort_values()
````

````python
to_downsample = 'BIO_ddseq_3'
min_value = df_stats['RPC'].sort_values()[to_downsample]
downsampling_depth = [min_value] * len(df_stats["RPC"]) / df_stats["RPC"]
````

````python
downsampling_depth = downsampling_depth.dropna()
````

````python
for item in downsampling_depth.index:
    if downsampling_depth[item] >= 1:
        downsampling_depth = downsampling_depth.drop(item)
````

````python
downsampling_depth
````

````python
import os
````

````python
parallel_filename = "downsample.libds.parallel"
with open(parallel_filename, 'w') as f:
    for sample in downsampling_depth.index:
        rate = downsampling_depth[sample]
        if rate <= 1.0:
            if sample.split('_')[1] == "ddseq":
                reads = [1, 2]
            else:
                reads = [1, 2, 3]

            for read in reads:
                fastq = f"full_fastq/{sample}__R{str(read)}.FULL.fastq.gz"
                newfastq = f"libds_fastq/{sample}__R{str(read)}.LIBDS.fastq.gz"
                if not os.path.exists(newfastq):
                    f.write(f"seqtk sample -s 111 {fastq} {rate} | bgzip -@ 8 > {newfastq}\n")
                    print(f"seqtk sample -s 111 {fastq} {rate} | bgzip -@ 8 > {newfastq}")

                else:
                    f.write(f"#{newfastq} already exists!\n")
                    print(f"#{newfastq} already exists!")
````

Then invoke `parallel` as such:

```
my_conda_initialize
module load HTSlib
cat downsample.libds.parallel | parallel -j 20 --progress
```

# Downsample scRNA-seq component of multiome

````python
glob.glob("full_fastq_scrna/*_lengths.txt")
````

````python
df = pd.read_csv(
    "full_fastq_scrna/R1_lengths.txt",
    sep="\t",
    index_col=0,
    header=None,
)
````

````python
df.index = [x.split("__")[0].replace("rna", "") for x in df.index]
````

````python
df.columns = ["reads"]
````

````python
data_obs_dict = {x.split("/")[-1].split("__")[0]: x for x in glob.glob("../full_5_cellranger/rna_qc/*.tsv")}
data_obs_dict
````

````python
for sample in df.index:
    print(sample)
    data_obs = pd.read_csv(data_obs_dict[sample], index_col=0, header=0, sep="\t")
    rpc = df.at[sample, "reads"] / len(data_obs)
    df.at[sample, "n_cells"] = len(data_obs)
    df.at[sample, "rpc"] = rpc
````

````python
df["rpc"]
````

````python
df["rpc"].idxmin()
````

````python
df["frac"] = [x / df["rpc"].min() for x in df["rpc"]]
df["frac_to_downsample"] = [df["rpc"].min() / x for x in df["rpc"]]
````

````python
df
````

````python
downsampling_depth = df["frac_to_downsample"].dropna()
downsampling_depth.drop(downsampling_depth[downsampling_depth == 1].index, inplace=True)
````

````python
downsampling_depth
````

````python
import os
parallel_filename = "downsample_multiome.libds.parallel"
with open(parallel_filename, 'w') as f:
    for sample in downsampling_depth.index:
        rate = downsampling_depth[sample]
        if rate <= 1.0:
            reads = [1, 2]

            for read in reads:
                fastq = f"full_fastq_scrna/{sample.replace('multiome','multiomerna')}__R{str(read)}.FULL.fastq.gz"
                newfastq = f"libds_fastq_scrna/{sample.replace('multiome','multiomerna')}__R{str(read)}.FULL.fastq.gz"
                if not os.path.exists(newfastq):
                    f.write(f"seqtk sample -s 111 {fastq} {rate} | bgzip -@ 8 > {newfastq}\n")
                    print(f"seqtk sample -s 111 {fastq} {rate} | bgzip -@ 8 > {newfastq}")

                else:
                    f.write(f"#{newfastq} already exists!\n")
                    print(f"#{newfastq} already exists!")
````
