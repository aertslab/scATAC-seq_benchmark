```python
with open(f'write_fragments.parallel', 'w') as f:
    for sample in bams_path_dict.keys():
        command = "create_fragments_file " + bams_path_dict[sample] + " " + sample + "_unused | coreutils sort --parallel=8 -S 16G -k 1,1V -k 2,2n -k 3,3n -k 4,4 | uniq -c | mawk -v 'OFS=\t' '{ print $2, $3, $4, $5, $1 }' | bgzip -@ 4 -c /dev/stdin > " + sample + ".fragments.raw.tsv.gz\n"
        f.write(command)
```
