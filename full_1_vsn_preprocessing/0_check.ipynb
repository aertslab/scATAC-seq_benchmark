```python
import glob
import pickle
import pandas as pd
import os
```

```python
cto_paths = sorted(glob.glob('cistopic_objects/*singlets*pkl'))
cto_paths
```

```python
for cto_path in cto_paths:
    cell_data_path = cto_path.replace(f'.pkl',f'.cell_data.tsv')
    if not os.path.exists(cell_data_path):
        print(cto_path)

        with open(cto_path, 'rb') as f:
            cto = pickle.load(f)
        
        cto.cell_data.to_csv(cell_data_path)
    else:
        print(f"{cell_data_path} exists")
```
