```python
import os
import glob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random
import pickle

%load_ext lab_black
```

```python
cell_data_path_dict = {
    x.split("/")[-1].split(".")[0]: x
    for x in sorted(
        glob.glob(
            "../fixedcells_3_cistopic_consensus/cistopic_objects/*singlets.cell_data.tsv"
        )
    )
}
```

```python
cto_path_dict = {
    x.split("/")[-1].split(".")[0]: x
    for x in sorted(
        glob.glob("../fixedcells_3_cistopic_consensus/cistopic_objects/*singlets.pkl")
    )
}
```

```python
df_merged = pd.DataFrame()
for sample, path in cell_data_path_dict.items():
    print(sample)
    df = pd.read_csv(path, index_col=0)
    df_merged = pd.concat([df_merged, df])
```

```python
for sample, path in cell_data_path_dict.items():
    print(sample)
    df = pd.read_csv(path, index_col=0)

    with open(cto_path_dict[sample], "rb") as f:
        cto_all = pickle.load(f)

    if len(df.index) >= 1000:
        bc_1k = random.sample(cto_all.cell_names, 1000)
        cto_1k = cto_all.subset(cells=bc_1k, copy=True)
        cto_path_new = "cistopic_objects_subsampled/" + cto_path_dict[sample].split(
            "/"
        )[-1].replace(".pkl", ".1k.pkl")

        with open(cto_path_new, "wb") as f:
            pickle.dump(cto_1k, f, protocol=4)

    if len(df.index) >= 2000:
        bc_2k = random.sample(cto_all.cell_names, 2000)
        cto_2k = cto_all.subset(cells=bc_2k, copy=True)
        cto_path_new = "cistopic_objects_subsampled/" + cto_path_dict[sample].split(
            "/"
        )[-1].replace(".pkl", ".2k.pkl")

        with open(cto_path_new, "wb") as f:
            pickle.dump(cto_2k, f, protocol=4)

    if len(df.index) >= 3000:
        bc_3k = random.sample(cto_all.cell_names, 3000)
        cto_3k = cto_all.subset(cells=bc_3k, copy=True)
        cto_path_new = "cistopic_objects_subsampled/" + cto_path_dict[sample].split(
            "/"
        )[-1].replace(".pkl", ".3k.pkl")

        with open(cto_path_new, "wb") as f:
            pickle.dump(cto_3k, f, protocol=4)
```
